### genre 추가 -  for문으로 500개 영화 넣어서 장르를 뽑아내자

```python
movies_genre = []
for movie in movies_title:
    try:
        driver.get('https://www.naver.com/')
        search_box = driver.find_element_by_id("query")
        search_box.send_keys("영화 " + str(movie))
        search_box.submit()
        driver.get(driver.find_element_by_class_name('sh_movie_link').get_attribute('href'))
        # BeautifulSoup이 가능하다면 이걸로 도전해보자
        
        new = driver.find_elements_by_css_selector('#content > div.article > div.mv_info_area > div.mv_info > dl > dd:nth-child(2) > p > span:nth-child(1)')
        words = new[0].text
        word = ''
        for i in range(len(words)):
            if words[i] != ",":
                word += words[i]
            else:
                break
        movies_genre.append(word)
        print(word)
    except:
        print(movie+' 안됨')
        movies_genre.append('')
```

```python
# 이제 뽑은 movies_title 리스트를 우리가 가진 csv파일의 'genre' 컬럼에 추가하자
movies['genre']=movies_genre
```

```python
# 그리고 저장하면 이 컬럼은 끝!
movies.to_csv('수정본.csv',index=False)
```

### 고화질 이미지 주소 가져오기

```python
from selenium import webdriver
import re
```

```python
driver=webdriver.Chrome('C:\\Users\\user\\Documents\\PythonStudy\\chromedriver.exe')
```

```python
driver.get('https://www.naver.com/')
```

```python
search_box = driver.find_element_by_id("query")
```

```python
search_box.send_keys("영화 " + "7번방의 선물")
```

```python
search_box.submit()
```

```python
driver.get(driver.find_element_by_class_name('sh_movie_link').get_attribute('href'))
```

```python
img = driver.find_element_by_xpath('//*[@id="content"]/div[1]/div[2]/div[2]/a/img')
```

```python
text = img.get_attribute("src")
```

```python
regex = re.compile(r'(http(s?):)([/|.|\w|\s|-])*\.(?:jpg|gif|png)')
```

```python
match = regex.search(text)
```

```python
poster_url = match.group()
poster_url
```

```python
print(poster_url)
```

```python
# 고화질 이미지 가져오기
movies_poster_url = []
num = 1
for movie in movies_title:
    try:
        driver.get('https://www.naver.com/')
        search_box = driver.find_element_by_id("query")
        search_box.send_keys("영화 " + str(movie))
        search_box.submit()
        driver.get(driver.find_element_by_class_name('sh_movie_link').get_attribute('href'))
        img = driver.find_element_by_xpath('//*[@id="content"]/div[1]/div[2]/div[2]/a/img')
        text = img.get_attribute("src")
        regex = re.compile(r'(http(s?):)([/|.|\w|\s|-])*\.(?:jpg|gif|png)')
        match = regex.search(text)
        poster_url = match.group()
        movies_poster_url.append(poster_url)
        print(num, poster_url)
        num += 1
    except:
        print(num, movie+' 안됨')
        num += 1
        movies_poster_url.append('')
```

```python
movies['new_poster_url'] = movies_poster_url
```

```python
movies.to_csv('movie_data(고화질포스터url추가).csv', index=False)
```
